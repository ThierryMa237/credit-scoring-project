```python
df = pd.read_csv('/Users/thierrymanuelzibiondobo/credit_scoring_project/data/raw/application_train.csv')

print(f"✅ Dataset chargé")
print(f"Dimensions : {df.shape[0]:,} lignes × {df.shape[1]} colonnes")
```

```python
df['DEBT_INCOME_RATIO'] = df['AMT_CREDIT'] / df['AMT_INCOME_TOTAL']
df['ANNUITY_INCOME_RATIO'] = df['AMT_ANNUITY'] / df['AMT_INCOME_TOTAL']
df['CREDIT_GOODS_RATIO'] = df['AMT_ANNUITY'] / df['AMT_GOODS_PRICE']
df['AGE_YEARS'] = df['DAYS_BIRTH'] / -365
df['YEARS_EMPLOYED'] = df['DAYS_EMPLOYED'] / -365
```

```python
#Vérification des valeurs

df[['DEBT_INCOME_RATIO', 'ANNUITY_INCOME_RATIO', 'CREDIT_GOODS_RATIO', 'AGE_YEARS', 'YEARS_EMPLOYED']].describe()
```

```python
df['YEARS_EMPLOYED'] = df['YEARS_EMPLOYED'].replace(df[df['YEARS_EMPLOYED'] < 0]['YEARS_EMPLOYED'].unique(), np.nan)

print(df['YEARS_EMPLOYED'].describe())
```

```python
df['CREDIT_TERM'] = df['AMT_CREDIT'] / df['AMT_ANNUITY']
df['INCOME_PER_PERSON'] = df['AMT_INCOME_TOTAL'] / df['CNT_FAM_MEMBERS']
df['EXT_SOURCE_MEAN'] = df[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']].mean(axis=1)
```

```python
df[['CREDIT_TERM', 'INCOME_PER_PERSON', 'EXT_SOURCE_MEAN']].describe()
```
